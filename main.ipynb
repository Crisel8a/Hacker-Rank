# Compare Triplets

```python
def compareTriplets(a, b):

  suma = 0
  sumb = 0
  for i in range(len(a)):
    if a[i] < b[i]:
      sumb += 1
    elif a[i] > b[i]:
      suma += 1
  print([suma, sumb])
```

```python
compareTriplets([5,6,7], [3,6,10])
```

# aVeryLongSum

```python
def aVeryLongSum(n, a):
  bigsum = 0
  if n == 10:
    print('no more than 10 elements')
    exit
  else:
    for i in range(len(a)):
      bigsum += a[i]
  return bigsum
```

```python
aVeryLongSum(5, [1000000001, 1000000002, 1000000003, 1000000004, 1000000005])
```

# Diagonal Difference

```python
def diagnalDifference(arr):
  diagonalDiference = 0
  for i in range(len(arr)):
    for j in range(len(arr)):
      if i == j:
        diagonalDiference += arr[i][j]
  return diagonalDiference
```
